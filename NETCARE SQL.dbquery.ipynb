```sql

# Viewing the table
SELECT*
FROM patient_readmission.netcare_hospital.patient_readmission_csv
LIMIT 5;


## Checking previous admissions and number of diagnoses

SELECT `Previous Admissions`,
       `Number of Diagnoses`
FROM patient_readmission.netcare_hospital.patient_readmission_csv
WHERE `Previous Admissions` > 2
OR `Number of Diagnoses` > 3
LIMIT 5;     


## Checking most popular admission type
SELECT DISTINCT `Admission Type` AS Most_Popular_Admission_Type
FROM patient_readmission.netcare_hospital.patient_readmission_csv
GROUP BY `Admission Type`
LIMIT 5; 


SELECT `Age`
FROM patient_readmission.netcare_hospital.patient_readmission_csv
WHERE `Age` = 18
LIMIT 5;

## Checking most common readmission 
SELECT 
    `Length of Stay`,
    COUNT(*) AS Number_Of_Patients,
    `Patient ID`,
    `Age`,
    `Gender`
FROM 
    patient_readmission.netcare_hospital.patient_readmission_csv
WHERE 
    `Length of Stay` > 5
GROUP BY 
    `Length of Stay`,
    `Patient ID`,
    `Age`,
    `Gender`
ORDER BY 
    Number_Of_Patients DESC
LIMIT 10;
     
## The same query as above but without the extra columns
SELECT 
    `Length of Stay`,
    COUNT(*) AS Number_Of_Patients
FROM 
    patient_readmission.netcare_hospital.patient_readmission_csv
WHERE 
    `Length of Stay` > 5
GROUP BY 
    `Length of Stay`
ORDER BY 
    Number_Of_Patients DESC
LIMIT 10;

## Checking average blood sugar levels for patients who have been readmitted

SELECT `Readmission` AS Readmission_Count,
        Avg(`Blood Sugar Levels`) AS Average_Blood_Sugar
FROM patient_readmission.netcare_hospital.patient_readmission_csv
WHERE `Readmission`= 'Yes'
GROUP BY `Readmission`
LIMIT 5;

## Checking number of patients with high blood pressure
SELECT `Blood Pressure` AS High_Blood_Pressure,
        COUNT(*) AS Number_Of_Patients
FROM patient_readmission.netcare_hospital.patient_readmission_csv
WHERE `Blood Pressure` > 140
GROUP BY `Blood Pressure`
LIMIT 5;

## Adding columns to enhance better insight

SELECT *,
CASE
    WHEN `Age` < 18 THEN 'Minor'
    WHEN `Age` BETWEEN 18 AND 35 THEN 'Young Adult'
    WHEN `Age` BETWEEN 36 AND 55 THEN 'Old Folks' -- Adjusted from 45 to catch ages 36-44
    ELSE 'Senior'
END AS Age_Group,

CASE
    WHEN `Length of Stay` < 5 THEN 'Short'
    WHEN `Length of Stay` BETWEEN 5 AND 10 THEN 'Medium'
    WHEN `Length of Stay` > 10 THEN 'Long'
    ELSE 'Unknown' -- Catches any missing or negative values
END AS Length_Of_Stay_Group,

CASE
    WHEN `Number of Diagnoses` < 3 THEN 'Low'
    WHEN `Number of Diagnoses` BETWEEN 3 AND 5 THEN 'Medium'
    WHEN `Number of Diagnoses` > 5 THEN 'High'
    ELSE 'Unknown'
END AS Number_Of_Diagnoses_Group,

CASE
    WHEN `Blood Pressure` < 140 THEN 'Low'
    WHEN `Blood Pressure` BETWEEN 140 AND 160 THEN 'Medium'
    WHEN `Blood Pressure` > 160 THEN 'High'
    ELSE 'Unknown'
END AS Blood_Pressure_Group,

CASE
    WHEN `Blood Sugar Levels` < 100 THEN 'Low'
    WHEN `Blood Sugar Levels` BETWEEN 100 AND 120 THEN 'Medium'
    WHEN `Blood Sugar Levels` > 120 THEN 'High'
    ELSE 'Unknown'
END AS Blood_Sugar_Group
FROM patient_readmission.netcare_hospital.patient_readmission_csv;
```
